```python
from google.colab import drive
import nltk  
import numpy as np  
import pandas as pd
import re 
```

```python
nltk.download('punkt')
```

```python
drive.mount('/content/drive')
```

```python
df = pd.read_excel('/content/drive/MyDrive/course_work/Comments.xlsx')

article_text = ''

for para in df['Comment']:  
    article_text += str(para)
```

```python
corpus = nltk.sent_tokenize(article_text)
corpus
```

```python
for i in range(len(corpus)):
    corpus [i] = corpus [i].lower()
    corpus [i] = re.sub(r'\W',' ',corpus [i])
    corpus [i] = re.sub(r'\s+',' ',corpus [i])
print(len(corpus))
```

```python
wordfreq = {}
for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
print(wordfreq)
```

```python
df_wordfreq = pd.DataFrame(wordfreq.items(), columns=['Word', 'Freq'])
df_wordfreq = df_wordfreq.sort_values('Freq', ascending=False)

df_wordfreq
```
